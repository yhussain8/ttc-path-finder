```python
stops_sample = stops.sample(100)
stops_sample = stops_sample.reset_index(drop = False)
stops_sample
```

```python
stops_map = folium.Map(location = [stops_sample.stop_lat.mean(), stops_sample.stop_lon.mean()], zoom_start = 12, width = 800, height = 600)
```

```python
for i in range(0, len(stops_sample)):
    marker_color = ''
    if (stops_sample.wheelchair_boarding[i] == 1):
        marker_color = 'green'
    else:
        marker_color = 'red'
    folium.Marker(
                    location = [stops_sample.stop_lat[i], stops_sample.stop_lon[i]],
                    popup = stops_sample.stop_name[i],
                    tooltip = stops_sample.stop_code[i],
                    icon = folium.Icon(color = marker_color, prefix = 'fa', icon = 'bus')
                 ).add_to(stops_map)
```

```python
fig = Figure(width = 800, height = 600)
fig.add_child(stops_map)
fig
```
